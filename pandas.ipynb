```python
import pandas as pd
import pandas_datareader.data as data
import yfinance as yfin
yfin.pdr_override()
df = data.get_data_yahoo('AAPL', start='2019-01-01', end='2020-01-01')
df['Close'] = df['Close'] * 1200 #  1200 
df['Close'].plot() #   plot   .

df2 = data.get_data_yahoo('005930.KS', start='2019-01-01', end='2020-01-01')
df2.index
df2['Close'].plot() #  plot   object    df2['Close'].astype(float)   
```

```python
#  .
df2['rolling5'] = df2['Close'].rolling(5).mean() # 5  .
df2['rolling20'] = df2['Close'].rolling(20).mean()
df2['rolling5'].plot()
df2['Close'].plot()
df2['rolling20'].plot()
```

```python
import matplotlib.pyplot as plt

plt.plot(df.index, df['Close'], color="red") # X,Y()   
plt.xlabel("time")
plt.ylabel("price")
plt.legend(['apple']) #  

plt.show() #   


plt.bar(df.index,df['Close'], color="skyblue") #    
plt.show()

plt.pie([10,20,50,20], labels=['drink','snack','ramen','desert']) #   
plt.show()

plt.hist([160,165, 170, 171,172,180]) #   
plt.show()

math  = [5, 8 , 34, 34, 11 , 23, 44, 55]
english  = [25, 84 , 4, 24, 43 , 13, 44, 65]
plt.scatter(math, english) #    
plt.show()

plt.stackplot([1,2,3],[10, 20, 30], [20, 40, 50]) #     ( X, Y1, Y2 ...)
plt.show()
```

```python
#  ( AMZN)  2020 0101~1231  yahoo   20, 60 
df3 = data.get_data_yahoo('AMZN', start='2020-01-01', end='2021-01-01')
df3['rolling20'] = df3['Close'].rolling(20).mean()
df3['rolling60'] = df3['Close'].rolling(60).mean()
plt.plot(df3.index, df3['rolling20'], color="red")
plt.plot(df3.index, df3['rolling60'], color="blue")
plt.plot(df3.index, df3['Close'], color='yellow')
plt.show()
```

```python
#   LG    
plt.plot([2018,2019,2020,2021],[50000,60000,75000,70000], color = "blue")
plt.plot([2018,2019,2020,2021],[30000,40000,50000,35000], color = "red")
plt.xlabel("year")
plt.ylabel("sales")
plt.legend(['Samsung','LG'])
plt.show()
```

```python
# 2020 0101 ~ 1231    
# , Volume 2020 Volume    .

df4 = data.get_data_yahoo('BTC-USD', start='2020-01-01', end='2021-01-01')
df4['high_volume_flag'] = df4['Volume'] > df4['Volume'].mean()
a = df4[df4['high_volume_flag']==True]
plt.bar(a.index, a['Close'])
plt.show()
```
