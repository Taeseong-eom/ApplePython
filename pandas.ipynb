```python
import pandas as pd
import pandas_datareader.data as data
import yfinance as yfin
yfin.pdr_override()
df = data.get_data_yahoo('AAPL', start='2019-01-01', end='2020-01-01')
df['Close'] = df['Close'] * 1200 # 평균환율 1200원 적용
df['Close'].plot() # 컬럼 하나에 plot 쓰면 그래프 보여줌.

df2 = data.get_data_yahoo('005930.KS', start='2019-01-01', end='2020-01-01')
df2.index
df2['Close'].plot() # 숫자데이터만 plot할 수 있는데 object 데이터로 되어 있으면 df2['Close'].astype(float) 하면 타입이 바뀜
```

```python
# 이동평균선을 그려보자.
df2['rolling5'] = df2['Close'].rolling(5).mean() # 5개씩 평균을 내줌.
df2['rolling20'] = df2['Close'].rolling(20).mean()
df2['rolling5'].plot()
df2['Close'].plot()
df2['rolling20'].plot()
```
