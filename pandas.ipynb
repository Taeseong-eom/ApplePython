```python
import pandas as pd
import pandas_datareader.data as data
import yfinance as yfin
yfin.pdr_override()
df = data.get_data_yahoo('AAPL', start='2019-01-01', end='2020-01-01')
df['Close'] = df['Close'] * 1200 # 평균환율 1200원 적용
df['Close'].plot() # 컬럼 하나에 plot 쓰면 그래프 보여줌.

df2 = data.get_data_yahoo('005930.KS', start='2019-01-01', end='2020-01-01')
df2.index
df2['Close'].plot() # 숫자데이터만 plot할 수 있는데 object 데이터로 되어 있으면 df2['Close'].astype(float) 하면 타입이 바뀜
```

```python
# 이동평균선을 그려보자.
df2['rolling5'] = df2['Close'].rolling(5).mean() # 5개씩 평균을 내줌.
df2['rolling20'] = df2['Close'].rolling(20).mean()
df2['rolling5'].plot()
df2['Close'].plot()
df2['rolling20'].plot()
```

```python
import matplotlib.pyplot as plt

plt.plot(df.index, df['Close'], color="red") # X,Y(리스트) 축 데이터를 설정
plt.xlabel("time")
plt.ylabel("price")
plt.legend(['apple']) # 범례 설정

plt.show() # 선 차트 출력


plt.bar(df.index,df['Close'], color="skyblue") # 데이터가 얼마 없을때 사용
plt.show()

plt.pie([10,20,50,20], labels=['drink','snack','ramen','desert']) # 비율 보여주고 싶을때
plt.show()

plt.hist([160,165, 170, 171,172,180]) # 분포가 어떻게 되는지
plt.show()

math  = [5, 8 , 34, 34, 11 , 23, 44, 55]
english  = [25, 84 , 4, 24, 43 , 13, 44, 65]
plt.scatter(math, english) # 데이터간 관계를 보고 싶을때
plt.show()

plt.stackplot([1,2,3],[10, 20, 30], [20, 40, 50]) # 누적 차트 보여주고 싶을때 ( X축, Y축1, Y축2 ...)
plt.show()
```

```python
plt.figure(figsize=(10,10)) # 인치 단위로 입력
plt.plot([1,2,3],[10, 20, 30])
plt.plot([1,2,3],[30, 10, 50])  # 여러 선을 동시에 보일 수 있음.

plt.show()
```

```python
# 아마존 (종목코드 AMZN) 주식의 2020년 01월01일~12월31일 주식가격을 yahoo에서 가져온 후 20일, 60일 이동평균선
df3 = data.get_data_yahoo('AMZN', start='2020-01-01', end='2021-01-01')
df3['rolling20'] = df3['Close'].rolling(20).mean()
df3['rolling60'] = df3['Close'].rolling(60).mean()
plt.plot(df3.index, df3['rolling20'], color="red")
plt.plot(df3.index, df3['rolling60'], color="blue")
plt.plot(df3.index, df3['Close'], color='yellow')
plt.show()
```

```python
# 삼성전자와 경쟁업체 LG전자의 매출 데이터를 표로 정리
plt.plot([2018,2019,2020,2021],[50000,60000,75000,70000], color = "blue")
plt.plot([2018,2019,2020,2021],[30000,40000,50000,35000], color = "red")
plt.xlabel("year")
plt.ylabel("sales")
plt.legend(['Samsung','LG'])
plt.show()
```

```python
# 2020년 01월01일 ~ 12월31일의 비트코인 가격을 그래프로 정리
# 단, Volume항목 2020년의 평균Volume보다 높은 날의 가격만 보여라.

df4 = data.get_data_yahoo('BTC-USD', start='2020-01-01', end='2021-01-01')
df4['high_volume_flag'] = df4['Volume'] > df4['Volume'].mean()
a = df4[df4['high_volume_flag']==True]
plt.bar(a.index, a['Close'])
plt.show()
```
